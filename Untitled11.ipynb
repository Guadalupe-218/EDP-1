```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
Lx, Ly = 4, 2  # Dimensiones del rectángulo
Nx, Ny = 50, 50  # Resolución espacial
Nt = 100  # Número de pasos en tiempo
c = 1.0  # Velocidad de onda
t_max = 10  # Tiempo máximo de simulación

x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
t = np.linspace(0, t_max, Nt)
X, Y = np.meshgrid(x, y)

# Función inicial
f_xy = 0.1 * (4*X - X*2) * (2*Y - Y*2)

# Parámetros de la serie
m_vals = np.arange(1, 10)
n_vals = np.arange(1, 10)

def series_solution(x, y, t, Lx, Ly, c):
    u_t = np.zeros_like(x)
    for m in m_vals:
        for n in n_vals:
            A_mn = 0.1  # Aproximación de coeficientes
            km = (m * np.pi / Lx)**2
            kn = (n * np.pi / Ly)**2
            u_t += A_mn * np.sin(m * np.pi * x / Lx) * np.sin(n * np.pi * y / Ly) * np.cos(c * np.sqrt(km + kn) * t)
    return u_t

# Simulación en un instante t
t_snapshot = 2.0  # Tomamos un instante de tiempo
u_xy_t = series_solution(X, Y, t_snapshot, Lx, Ly, c)

# Visualización
plt.figure(figsize=(8,6))
plt.contourf(X, Y, u_xy_t, cmap='viridis')
plt.colorbar(label='Amplitud')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Solución de la ecuación de onda en t={t_snapshot}')
```
