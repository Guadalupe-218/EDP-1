```python

import numpy as np
import matplotlib.pyplot as plt
import sympy as sp

# Definimos la funcin por partes
def F(x):
    if -2 <= x < 0:
        return 0
    elif 0 <= x <= 2:
        return 2 - x
    else:
        return 0

# Dominio
x = np.linspace(-2, 2, 1000)
y = np.array([F(i) for i in x])

# Graficamos la funcin original
plt.plot(x, y, label="F(x)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Funcin F(x) por partes")
plt.legend()
plt.grid()
plt.show()

# Calculamos la serie de Fourier
x_sym = sp.Symbol('x')
f_sym = sp.Piecewise((0, (x_sym >= -2) & (x_sym < 0)),
                     (2 - x_sym, (x_sym >= 0) & (x_sym <= 2)))
n = 10  # Nmero de trminos en la serie de Fourier

a0 = (1/4) * sp.integrate(f_sym, (x_sym, -2, 2))
an = [(1/2) * sp.integrate(f_sym * sp.cos(n*sp.pi*x_sym/2), (x_sym, -2, 2)) for n in range(1, n+1)]
bn = [(1/2) * sp.integrate(f_sym * sp.sin(n*sp.pi*x_sym/2), (x_sym, -2, 2)) for n in range(1, n+1)]

# Evaluamos la serie de Fourier
F_fourier = a0
for i in range(n):
    F_fourier += an[i] * sp.cos((i+1) * sp.pi * x_sym / 2) + bn[i] * sp.sin((i+1) * sp.pi * x_sym / 2)

# Convertimos la serie de Fourier en una funcin evaluable
F_fourier_func = sp.lambdify(x_sym, F_fourier, 'numpy')

# Graficamos la serie de Fourier junto con la funcin original
y_fourier = F_fourier_func(x)
plt.plot(x, y, label="F(x)")
plt.plot(x, y_fourier, label="Serie de Fourier")
plt.xlabel("x")
plt.ylabel("F(x) y Serie de Fourier")
plt.title("Funcin F(x) y su Serie de Fourier")
plt.legend()
plt.grid()
plt.show()
```
