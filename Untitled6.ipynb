```python
import sympy as sp

# Definir variables
x = sp.symbols('x')
lamda = sp.symbols('lambda')
f = sp.Function('f')

# Definir la ecuación diferencial
eq = sp.Eq(sp.diff(f(x), x, x) + 2*sp.diff(f(x), x) + lamda*f(x), 0)

# Resolver la ecuación diferencial
sol_general = sp.dsolve(eq, f(x))
print("Solución general:", sol_general)
```

```python
# Definir las constantes de integración
C1, C2 = sp.symbols('C1 C2')

# Expresar la solución en términos de constantes
sol_expr = sol_general.rhs.subs({sp.symbols('C1'): C1, sp.symbols('C2'): C2})

# Aplicar condiciones de frontera
conds = [sol_expr.subs(x, 0), sol_expr.subs(x, 1)]
sol_constants = sp.solve(conds, [C1, C2])

# Sustituir las constantes en la solución final
sol_final = sol_expr.subs(sol_constants)
print("Solución con condiciones de frontera:", sol_final)
```

```python
# Definir los valores propios
n = sp.symbols('n', integer=True)
lambda_n = (n * sp.pi)**2
print("Valores propios de lambda:", lambda_n)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Convertir la función propia a una función numérica
f_n_expr = sp.exp(-x) * sp.sin(n * sp.pi * x)
f_n = sp.lambdify((x, n), f_n_expr, 'numpy')

# Generar valores de x
x_vals = np.linspace(0, 1, 100)

# Graficar algunas funciones propias
plt.figure(figsize=(8, 6))
for i in range(1, 4):  # Graficamos para n = 1, 2, 3
    y_vals = f_n(x_vals, i)
    plt.plot(x_vals, y_vals, label=f'n={i}')

plt.xlabel('x')
plt.ylabel('f_n(x)')
plt.title('Funciones propias del problema de Sturm-Liouville')
plt.legend()
plt.grid()
plt.show()
```
