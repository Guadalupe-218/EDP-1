```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

# Valores de x en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

# Crear la figura
plt.figure(figsize=(8, 6))

# Graficar los polinomios de Legendre
for n in range(6):
    Pn = legendre(n)
    plt.plot(x, Pn(x), label=f'P_{n}(x)')

# Configuración de la gráfica
plt.xlabel('x')
plt.ylabel('P_n(x)')
plt.title('Polinomios de Legendre')
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import legendre

# Definir n
n = 3

# Valores de x en el intervalo [-1, 1]
x = np.linspace(-1, 1, 100)

# Polinomios de Legendre P_n(x) y P_(n+1)(x)
Pn = legendre(n)
Pn1 = legendre(n+1)

# Derivada de P_n(x)
Pn_derivada = np.polyder(Pn)

# Calcular ambas expresiones de la ecuación
lado_izq = (1 - x**2) * np.polyval(Pn_derivada, x)
lado_der = (n + 1) * x * Pn(x) - (n + 1) * Pn1(x)

# Graficar
plt.figure(figsize=(8, 6))
plt.plot(x, lado_izq, label=r'$(1-x^2) P_n^{\prime}(x)$')
plt.plot(x, lado_der, label=r'$(n+1)x P_n(x) - (n+1) P_{n+1}(x)$', linestyle='dashed')
plt.xlabel('x')
plt.ylabel('Valor')
plt.title(f'Verificación gráfica de la relación de recurrencia para n={n}')
plt.legend()
plt.grid()
plt.show()
```



Demostración de la relación de recurrencia para los polinomios de Legendre


Queremos demostrar algebraicamente que se satisface la relación:

(1-x^2) P'n (x) = (n + 1)x P_n (x) - (n+1) P{n+1} (x)
\end{equation}

para \( n = 3 \).


Los polinomios de Legendre se definen como:

\begin{equation}
P_n(x) = \frac{1}{2^n} \sum_{k=0}^{\lfloor n/2 \rfloor} (-1)^k \frac{(2n - 2k)!}{k!(n-k)!(n-2k)!} x^{n-2k}
\end{equation}

Para \( n = 3 \), el polinomio de Legendre es:

\begin{equation}
P_3(x) = \frac{1}{2} (5x^3 - 3x)
\end{equation}

El siguiente polinomio es:

\begin{equation}
P_4(x) = \frac{1}{8} (35x^4 - 30x^2 + 3)
\end{equation}

\section{Cálculo de la derivada}

La derivada de \( P_3(x) \) es:

\begin{equation}
P'_3(x) = \frac{d}{dx} \left( \frac{1}{2} (5x^3 - 3x) \right) = \frac{1}{2} (15x^2 - 3)
\end{equation}

Multiplicamos por \( (1 - x^2) \):

\begin{equation}
(1 - x^2) P'_3(x) = (1 - x^2) \frac{1}{2} (15x^2 - 3)
\end{equation}

Distribuyendo,

\begin{equation}
= \frac{1}{2} (15x^2 - 3 - 15x^4 + 3x^2)
\end{equation}

\begin{equation}
= \frac{1}{2} (-15x^4 + 18x^2 - 3)
\end{equation}

\section{Cálculo del lado derecho}

\begin{equation}
4x P_3(x) = 4x \cdot \frac{1}{2} (5x^3 - 3x) = 10x^4 - 6x^2
\end{equation}

\begin{equation}
4 P_4(x) = 4 \cdot \frac{1}{8} (35x^4 - 30x^2 + 3) = \frac{4}{8} (35x^4 - 30x^2 + 3)
\end{equation}

\begin{equation}
= \frac{1}{2} (35x^4 - 30x^2 + 3)
\end{equation}

Restamos:

\begin{equation}
(10x^4 - 6x^2) - \frac{1}{2} (35x^4 - 30x^2 + 3)
\end{equation}

Distribuyendo el \( \frac{1}{2} \):

\begin{equation}
10x^4 - 6x^2 - \frac{35}{2}x^4 + \frac{30}{2}x^2 - \frac{3}{2}
\end{equation}

\begin{equation}
= -\frac{15}{2}x^4 + \frac{18}{2}x^2 - \frac{3}{2}
\end{equation}

\begin{equation}
= \frac{1}{2} (-15x^4 + 18x^2 - 3)
\end{equation}


Dado que ambos lados de la ecuación son iguales, hemos demostrado algebraicamente que:

(1-x^2) P'_3 (x) = 4x P_3 (x) - 4 P_4 (x)
\end{equation}
