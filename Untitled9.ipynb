```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x'
x_prime = np.linspace(0, 1, 1000)

# Definir la función f(x')
f_x_prime = np.exp(-((x_prime - 0.5)**2) / 0.01)

# Graficar la función
plt.plot(x_prime, f_x_prime, label=r'$f(x\') = e^{-\frac{(x\' - 0.5)^2}{0.01}}$')
plt.xlabel("x'")
plt.ylabel("f(x')")
plt.title("Perturbación en una cuerda")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(0, 2, 1000)  # Aumentamos el rango para ver el desplazamiento
c = 1  # Velocidad de propagación
t = 0.5  # Momento en el tiempo

# Cambio de variable x' = x - ct
x_prime = x - c*t

# Definir la función f(x') con el cambio de variable
f_x_prime = np.exp(-((x_prime - 0.5)**2) / 0.01)

# Graficar la función
plt.plot(x, f_x_prime, label=r"$f(x') = e^{-\frac{(x - ct - 0.5)^2}{0.01}}$")
plt.xlabel("x")
plt.ylabel("f(x')")
plt.title("Desplazamiento de la perturbación hacia la derecha")
plt.legend()
plt.grid(True)
plt.show()
```

 La función se mantiene en la misma forma, pero ahora su argumento depende explícitamente de t .A medida que aumenta  t , el término -ct  hace que la perturbación se desplace hacia valores más grandes de  x , es decir, hacia la derecha.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(-1, 1, 1000)  # Ajustamos el rango para visualizar el desplazamiento
c = 1  # Velocidad de propagación
t = 0.5  # Momento en el tiempo

# Cambio de variable x' = x + ct
x_prime = x + c*t

# Definir la función f(x') con el cambio de variable
f_x_prime = np.exp(-((x_prime - 0.5)**2) / 0.01)

# Graficar la función
plt.plot(x, f_x_prime, label=r"$f(x') = e^{-\frac{(x + ct - 0.5)^2}{0.01}}$")
plt.xlabel("x")
plt.ylabel("f(x')")
plt.title("Desplazamiento de la perturbación hacia la izquierda")
plt.legend()
plt.grid(True)
plt.show()
```

 Al modificar la variable con \( x' = x + ct \), la perturbación se mueve hacia la izquierda con el tiempo. La estructura de la onda permanece igual, pero su posición cambia.La gráfica muestra la evolución en \( x \), destacando el desplazamiento hacia valores más pequeños.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(-1, 2, 1000)  # Ajustamos el rango para visualizar ambas ondas
c = 1  # Velocidad de propagación
t = 0.5  # Momento en el tiempo

# Definir las perturbaciones individuales
x_prime_right = x - c*t  # Onda viajera hacia la derecha
x_prime_left = x + c*t   # Onda viajera hacia la izquierda

f_right = np.exp(-((x_prime_right - 0.5)**2) / 0.01)
f_left = np.exp(-((x_prime_left - 0.5)**2) / 0.01)

# Superposición de las dos ondas
f_superposition = f_right + f_left

# Graficar las ondas individuales y la superposición
plt.plot(x, f_right, '--', label="Onda viajera a la derecha", alpha=0.7)
plt.plot(x, f_left, '--', label="Onda viajera a la izquierda", alpha=0.7)
plt.plot(x, f_superposition, label="Superposición", linewidth=2)

# Configuración de la gráfica
plt.xlabel("x")
plt.ylabel("f(x')")
plt.title("Superposición de ondas viajeras")
plt.legend()
plt.grid(True)
plt.show()
```
