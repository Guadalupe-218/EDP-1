```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir las variables simbólicas
x, w = sp.symbols('x w')

# Definir la función pulso unitario
f = sp.Piecewise((1, (-1 <= x) & (x <= 1)), (0, True))

# Calcular la Transformada de Fourier como una integral simbólica
F_w = sp.integrate(f * sp.exp(-1j * w * x), (x, -1, 1))

# Mostrar la integral calculada
print("Transformada de Fourier:")
sp.pprint(F_w)

# Aproximación con Serie de Fourier truncada
N = 10  # Número de términos en la aproximación
fourier_series = sum((2 * sp.sin(n * sp.pi) / (n * sp.pi)) * sp.cos(n * x) for n in range(1, N + 1))

# Crear la figura para graficar
x_vals = np.linspace(-2, 2, 400)
f_numeric = np.where((x_vals >= -1) & (x_vals <= 1), 1, 0)
fourier_numeric = np.array([fourier_series.subs(x, val).evalf() for val in x_vals])

plt.figure(figsize=(8, 5))
plt.plot(x_vals, f_numeric, label="Pulso Unitario", linewidth=2, color='black')
plt.plot(x_vals, fourier_numeric, '--', label="Serie de Fourier truncada (N=10)", linewidth=2, color='blue')
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Comparación entre Función Original y Serie de Fourier")
plt.grid()
plt.show()
```

```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

# Definir las variables simbólicas
x, w, a = sp.symbols('x w a')

# Definir la función pulso unitario
f = sp.Piecewise((1, (-1 <= x) & (x <= 1)), (0, True))

# Aproximación a la transformada de Fourier con límite superior finito
F_w_a = sp.integrate(f * sp.exp(-1j * w * x), (x, -a, a))

# Convertir la expresión simbólica en una función evaluable
F_w_a_func = sp.lambdify((w, a), F_w_a, "numpy")

# Definir valores de 'a'
A_vals = [8, 16, 32]

# Generar valores de frecuencia w para evaluar la integral
w_vals = np.linspace(-10, 10, 400)

# Graficar cada aproximación en su propio panel
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)

for i, val in enumerate(A_vals):
    F_numeric = F_w_a_func(w_vals, val).real  # Evaluar la función numéricamente
    axs[i].plot(w_vals, F_numeric, label=f"Integral truncada a={val}", linewidth=2, color='blue')
    axs[i].legend()
    axs[i].set_ylabel("F(w)")
    axs[i].grid()
    axs[i].set_title(f"Integral de Fourier truncada para a={val}")

axs[-1].set_xlabel("w")
plt.tight_layout()
plt.show()
```
