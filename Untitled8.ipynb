```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros
L = 1  # Longitud de la cuerda
c = 1  # Velocidad de propagacin
K = 1  # Amplitud mxima

# Nmero de trminos en la serie de Fourier
N = 50
x = np.linspace(0, L, 100)  # Puntos en la cuerda
t_vals = np.linspace(0, 2, 5)  # Valores de tiempo

# Funcin f(x) (deflexin inicial triangular)
def f(x, L, K):
    return np.where(x < L/2, (2*K*x)/L, (2*K*(L-x))/L)

# Expansin en serie de Fourier
def u(x, t, L, c, K, N):
    suma = np.zeros_like(x)
    for n in range(1, N+1):
        lambda_n = n * np.pi / L
        B_n = (4*K / (n*2 * np.pi*2)) * np.sin(n*np.pi/2)
        suma += B_n * np.sin(lambda_n * x) * np.cos(c * lambda_n * t)
    return suma

# Graficar resultados en distintos tiempos
plt.figure(figsize=(8, 6))
for t in t_vals:
    plt.plot(x, u(x, t, L, c, K, N), label=f't={t:.2f}')

plt.xlabel('Posicin x')
plt.ylabel('Desplazamiento u(x,t)')
plt.title('Cuerda Vibratoria con Deflexin Inicial Triangular')
plt.legend()
plt.grid()
plt.show()
```

Para resolver la ecuacin diferencial \( X'' -  X = 0 \) con condiciones de frontera \( X(0) = 0 \) y \( X(1) = 0 \), podemos encontrar los valores propios y funciones propias utilizando el mtodo de separacin de variables.

La ecuacin caracterstica es:

\[
X'' -  X = 0
\]

La solucin general depende del signo de \(  \):

1. Si \(  = ^2 > 0 \), la solucin es \( X(x) = A cos( x) + B sin( x) \).
2. Aplicando las condiciones de frontera:
   - \( X(0) = 0  A = 0 \), por lo que \( X(x) = B \sin(\mu x) \).
   - \( X(1) = 0  B \sin() = 0 \), lo que implica que \(  = n \) para \( n = 1,2,3,  \).

Por lo tanto, las soluciones propias son:

\[
X_n(x) = sin(n x)
\]

y los valores propios son:

\[
()_n = (n)^2
\]


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir dominio
x = np.linspace(0, 1, 100)

# Nmero de funciones propias a visualizar
N = 5

plt.figure(figsize=(8, 6))
for n in range(1, N+1):
    X_n = np.sin(n * np.pi * x)  # Solucin X_n(x)
    plt.plot(x, X_n, label=f'n={n}')

plt.xlabel('x')
plt.ylabel('X_n(x)')
plt.title('Funciones propias de la ecuacin diferencial')
plt.legend()
plt.grid()
plt.show()
```
