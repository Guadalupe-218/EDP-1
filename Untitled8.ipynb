```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
L = 1  # Longitud de la cuerda
c = 1  # Velocidad de propagación
K = 1  # Amplitud máxima

# Número de términos en la serie de Fourier
N = 50
x = np.linspace(0, L, 100)  # Puntos en la cuerda
t_vals = np.linspace(0, 2, 5)  # Valores de tiempo

# Función f(x) (deflexión inicial triangular)
def f(x, L, K):
    return np.where(x < L/2, (2*K*x)/L, (2*K*(L-x))/L)

# Expansión en serie de Fourier
def u(x, t, L, c, K, N):
    suma = np.zeros_like(x)
    for n in range(1, N+1):
        lambda_n = n * np.pi / L
        B_n = (4*K / (n*2 * np.pi*2)) * np.sin(n*np.pi/2)
        suma += B_n * np.sin(lambda_n * x) * np.cos(c * lambda_n * t)
    return suma

# Graficar resultados en distintos tiempos
plt.figure(figsize=(8, 6))
for t in t_vals:
    plt.plot(x, u(x, t, L, c, K, N), label=f't={t:.2f}')

plt.xlabel('Posición x')
plt.ylabel('Desplazamiento u(x,t)')
plt.title('Cuerda Vibratoria con Deflexión Inicial Triangular')
plt.legend()
plt.grid()
plt.show()
```

Para resolver la ecuación diferencial \( X'' - λ X = 0 \) con condiciones de frontera \( X(0) = 0 \) y \( X(1) = 0 \), podemos encontrar los valores propios y funciones propias utilizando el método de separación de variables.

La ecuación característica es:

\[
X'' - λ X = 0
\]

La solución general depende del signo de \( λ \):

1. Si \( λ = μ^2 > 0 \), la solución es \( X(x) = A cos(μ x) + B sin(μ x) \).
2. Aplicando las condiciones de frontera:
   - \( X(0) = 0 ⇒ A = 0 \), por lo que \( X(x) = B \sin(\mu x) \).
   - \( X(1) = 0 ⇒ B \sin(μ) = 0 \), lo que implica que \( μ = nπ \) para \( n = 1,2,3, ⋯ \).

Por lo tanto, las soluciones propias son:

\[
X_n(x) = sin(nπ x)
\]

y los valores propios son:

\[
(λ)_n = (nπ)^2
\]


```python
import numpy as np
import matplotlib.pyplot as plt

# Definir dominio
x = np.linspace(0, 1, 100)

# Número de funciones propias a visualizar
N = 5

plt.figure(figsize=(8, 6))
for n in range(1, N+1):
    X_n = np.sin(n * np.pi * x)  # Solución X_n(x)
    plt.plot(x, X_n, label=f'n={n}')

plt.xlabel('x')
plt.ylabel('X_n(x)')
plt.title('Funciones propias de la ecuación diferencial')
plt.legend()
plt.grid()
plt.show()
```
